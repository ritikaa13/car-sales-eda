#### **Domain Knowledge**

**Price:** The selling price of the car, which depends on various factors like brand, model, mileage, engine type, and year of manufacturing

**Body Type:** The structure of the car (e.g., sedan, hatchback, SUV, coupe, convertible, etc.), which affects market demand.

**Mileage:** In the context of cars in Ukraine, "mileage" refers to the total distance a vehicle has traveled since it was new, typically measured in kilometers (km). and unit for this column is thousand kilometers (e.g. 68 = 68000KM)

“Mileage” refers to the average distance or number of kilometers your car can go on a given quantity of fuel.

**Engine Volume (engV):** The engine displacement in liters (L). Higher engine volume usually means higher power but lower fuel efficiency.

**Engine Type (engType):** Fuel type of the engine: Petrol (Gasoline) – Common, high availability, and moderate cost. Diesel – Better mileage but higher maintenance and environmental concerns. Hybrid – A mix of fuel and electric power, improving fuel efficiency. Electric (EV) – Zero emissions and low running cost but higher initial price. LPG (Liquefied Petroleum Gas) – Alternative fuel that is cheaper but less common.

**Registration:** Whether the car has a valid registration. Registered cars have legal clearance and are easier to sell. Unregistered cars may require additional paperwork and legal checks.

**Year:** The year the car was manufactured. Older cars tend to be cheaper due to depreciation.

**Model:** The specific version of the car from the manufacturer.

Some models hold value better due to reliability and brand reputation. Drive Type = Full (All-Wheel Drive) = Power is delivered to all four wheels automatically, adjusting based on road conditions. Rear (Rear-Wheel Drive) = Power is delivered to the rear wheels, while the front wheels handle steering. Front (Front-Wheel Drive) = Power is delivered to the front wheels.

**Data Visualization(EDA)**

EDA is the first step in data analysis where we explore, clean, and summarize data to extract meaningful insights.

```python
# 4.1 - Which type of cars are sold maximum?
# Note - For frequency check we use histogram and count plot 

sns.countplot(data =df, x ="body")
plt.show()
```

**Answer -** In ukrain sedan car are sold max.

```python
# 4.2 What is the co-relation between price and mileage?

sns.scatterplot(data=df,x="mileage",y="price")
plt.show()
```

```python
sns.regplot(data = df,x="mileage",y="price")
plt.show()
```

**Answer -** There is a negative correlation between price and mileage. As mileage increases, the price of the car tends to decrease.

```python
# 4.3 How many cars are registered?

ax = sns.countplot(data=df,x="registration")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

**Answer -** The number of registered cars is significantly higher than the number of non-registered cars. Most cars in the dataset are registered.

```python
# 4.4 Price distribution between registered and non-registered cars.

sns.boxplot(data=df, x="price", y="registration")
plt.show()
```

**Answer -** Registered cars are generally priced higher than non-registered cars, with a wider and higher price distribution.

```python
# 4.5 What is the car price distribution based on Engine Value?

sns.scatterplot(data=df,x="engV",y="price")
plt.show()
```

**Answer -** There is a weak or unclear relationship between engine value and price, with high variability and some outliers present.
